```
import requests
import json
import sys
import csv

x = requests.get('https://covid19-api.org/api/timeline/ID')
resp = json.loads(x.text)

# data = resp[0]['cases']

# csv_writer = csv.writer()
#convert to csv
with open('coba.csv', 'w', newline='') as f:
  thewriter = csv.writer(f)

  thewriter.writerow(['country','last_update', 'cases', 'deaths', 'recovered'])

  for i in range(len(resp)):
  # print(len(resp))
    thewriter.writerow([resp[i]['country'], resp[i]['last_update'], resp[i]['cases'], resp[i]['deaths'], resp[i]['recovered']]) 
    print ("country")
    print(resp[i]['country'])
    print("last_update")
    print(resp[i]['last_update'])
    print("cases")
    print(resp[i]['cases'])
    print("deaths")
    print(resp[i]['deaths'])
    print("recovered")
    print(resp[i]['recovered'])

  print(x.text)
```
